I am going to analyze the data of a retail company, and attempt to understand the behavior of their customers. I want to find some interesting insights to plan their next campaigns, the target of a new product, aho are the most valuable customers, etc.

The original dataset, that can be found at: https://www.kaggle.com/datasets/imakash3011/customer-personality-analysis. It contains information of 2240 customers, such as their age, marital status, number of kids, education, income, number of purchases per channel, total spent per item, how many purchases had discount, whether they accepted their campaig noffers, etc. We have 29 attributes, that have a complete description on the original source.

##Clustering

###Standard Scaler

Some of the transformations on our data (PCA and Clustering) will involve measuring distances. That can be problematic, considering that some features have very different ranges. We will use StandardScaler, so they can all have a similar range (mostly between -1 to 1).

```python
#Standardization
sc = StandardScaler()
data_scaled = pd.DataFrame(sc.fit_transform(data), columns = data.columns)
display(data_scaled[:5].T)

```

###
Principal Components Analysis (PCA)

In the dataset, we have a high number of features and a lot of them have a certain level of correlation between them, and are somewhat redundant. Performing a dimensionality reduction technique is a good option to reduce the number of features, decreasing the computation cost of the algorithm, while maintaining a good amount of information from the original dataset.

This step is not mandatory, and is not really clear whether to use PCA before clustering really gives us better results. That is particularly tricky, since clustering is a unsupervised algorithm, which makes it harder to assess its performance.

Now, performing the PCA, we need to determine how many principal components we'll need to maintain about 70%~80% of variance of the original data.


```python
# Determining the number of principal components

variance_ratio = {}
for i in range(1, len(data_scaled.columns)+1):
    pca = PCA(n_components=i)
    pca.fit(data_scaled)
    variance_ratio[f'n_{i}'] = pca.explained_variance_ratio_.sum()*100
  
```

With 4 principal components, we have almost 70% of the variance explained. That's a good result.

```python
# Applying PCA to create a new dataset with 4 Princpal Components.

pca = PCA(n_components = 4, random_state = 123)
pca.fit(data_scaled)
data_pca = pd.DataFrame(pca.transform(data_scaled),
                      columns = (["PC1", "PC2", "PC3", "PC4"]))
data_pca.head(10)

```

Now we have a new dataset with the 4 principal components. In the next step, we are going to apply the k-means on this new dataset.

#
K-Means

We are going to use the k-means algorithm to create our clusters. First, we'll need to determine the ideal number of clusters. We are going to do it by comparing two techniques: the elbow method and the silhouette score.

In the elbow method, we plot the WCSS vs the number of clusters. The WCSS is the sum of squared distance between each point and the centroid in a cluster. Naturally, as we increase the number of clusters, this value will always get smaller.

What we are looking for is a value of N that, from that point forward, the decrease of WCSS won't be very significative.



```python
# Determining the ideal number of clusters: Elbow method
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=123)
    kmeans.fit(data_pca)
    wcss.append(kmeans.inertia_)
    
plt.figure(figsize=(10,6))
plt.plot(range(1,11), wcss, marker='o', color = 'tab:blue')
plt.title('Elbow Method', fontsize = 14, fontweight = 'bold')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```



Taking a look at the curve, it's not really clear what is the ideal number of clusters. Let's double check it with the silhouette scores. From wikipedia: The silhouette value is a measure of how similar an object is to its own cluster (cohesion) compared to other clusters (separation).


```python
silhouette_scores = []

for i in range(2,10):
    m1 = KMeans(n_clusters=i, random_state=123)
    c = m1.fit_predict(data_pca)
    silhouette_scores.append(silhouette_score(data_pca, m1.fit_predict(data_pca)))

plt.figure(figsize = (8,5))
plt.bar(range(2,10), silhouette_scores, color = 'darkgreen')
plt.title("Silhouette Scores vs Number of Clusters", fontsize = 14, fontweight = 'bold')
plt.xlabel('Number of clusters', fontsize = 14) 
plt.ylabel('Silhouette Scores', fontsize = 14)
plt.ylim(0,0.35)
for a, b in enumerate(silhouette_scores):
    plt.text(a + 1.7, b + 0.005, str(round(b,3)), color='black', fontweight='bold')
plt.show()
```

We are going to use K-Means with 3 clusters, and add the results to the original dataset.

“The silhouette value is a measure of how similar an object is to its own cluster (cohesion) compared to other clusters (separation).”

```python
kmeans = KMeans(n_clusters=3, random_state=123).fit(data_pca)
pred = kmeans.predict(data_pca)
data['cluster'] = pred + 1
data.head()
```
